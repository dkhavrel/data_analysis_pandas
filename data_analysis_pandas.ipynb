```python
wine["cost_per_point"] = wine["price"]/wine["points"]
```

```python
def findyear(x):
    for i in x.split(" "):
        if i.isdigit() and len(i) == 4 and int(i) > 1950 and int(i) < 2020:
            return int(i)
    return np.nan
```

```python
# Opdracht 16
```

```python
# Opdracht 17

wine["year"] = wine["title"].apply(findyear)
wine.describe()
```
